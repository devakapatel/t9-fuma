#Perform a practical KNN algorithm.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load a sample dataset (e.g., the Iris dataset)
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Create a KNN classifier with n_neighbors=3
knn = KNeighborsClassifier(n_neighbors=3)

# Train the model using the training data
knn.fit(X_train, y_train)

# Predict on the test data
y_pred = knn.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy
print(f"Accuracy: {accuracy:.2f}")
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Since the iris dataset has more than 2 features, we'll use PCA to reduce
# the dimensionality for visualization purposes.
pca = PCA(n_components=2)
X_test_reduced = pca.fit_transform(X_test)

# Create a scatter plot of the test data points, colored by predicted class
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_test_reduced[:, 0], X_test_reduced[:, 1], c=y_pred, cmap='viridis')
plt.title('KNN Classification Results (Test Data)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(*scatter.legend_elements(), title="Classes")
plt.colorbar(scatter, label='Predicted Class')
plt.show()
```
