```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the dataset
iris = load_iris()
X = iris.data
y = iris.target

# Convert to a pandas DataFrame for easier handling
df = pd.DataFrame(X, columns=iris.feature_names)
df['target'] = y

# For binary classification, let's consider only two classes (e.g., class 0 and class 1)
df_binary = df[df['target'].isin([0, 1])]

X_binary = df_binary.drop('target', axis=1)
y_binary = df_binary['target']

print("Dataset loaded successfully.")
print("Shape of the dataset:", df_binary.shape)
print("First 5 rows of the dataset:")
display(df_binary.head())
```
